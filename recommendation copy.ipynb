```python
# package
import pandas as pd
import numpy as np
from numpy import dot
from numpy.linalg import norm
import pickle
from sentence_transformers import SentenceTransformer
```

```python
# pre-trained model
save_directory = "./save_pretrained"
model = SentenceTransformer(save_directory)

# cosine similarity
def cos_sim(A, B):
  return dot(A, B)/(norm(A)*norm(B))
```

```python
# ë¦¬ë·° dataframe ë¶ˆëŸ¬ì˜¤ê¸°
fname = 'reviews.pkl'
file = open(fname, 'rb')
df = pickle.load(file)
file.close()
df
```

```python
# ìœ„ì¹˜ì™€ ë©”ë‰´ë¡œ í•„í„°ë§
loc = 'ì—­ì‚¼ë™'
menu = 'ë¹™ìˆ˜'
df = df.loc[df['dong'] == loc]
df['string'] = df['name'] + df['cate_1'] + df['cate_2'] + df['cate_3'] + df['type'] + df['review']
df= df[df['string'].apply(lambda x: menu in x)]
```

```python
# í•„í„°ë§ëœ dfì™€ í‚¤ì›Œë“œë¡œ ìœ ì‚¬ë„ ê¸°ë°˜ ì¶”ì²œ
def recommend(df, query):
  embedding = model.encode(query)
  # df = df.copy()
  df['score'] = df.apply(lambda x: cos_sim(x['embedding'], embedding), axis=1)
  result = df.loc[df['score'].idxmax()]
  return result['name'], result['naverURL']
```

```python
# ì‹ë‹¹ëª…, ë§í¬ ì¶œ
name, url = recommend(df, 'ë¶€ëª¨ë‹˜ ëª¨ì‹œê³  ê°ˆë§Œí•œ ì‹ë‹¹')
print(f"[ì§€ì—­]ì˜ [ìŒì‹] ë§›ì§‘ìœ¼ë¡œ {name} ì–´ë– ì„¸ìš”? \në§í¬: {url}")
# ì§€ì—­, ìŒì‹ì€ ì…ë ¥ê°’ìœ¼ë¡œ
```
